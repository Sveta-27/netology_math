```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
from scipy.stats import rv_continuous
```

Создайте случайную величину из любого выбранного вами распределения, 
для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
# gaussian with sigma = 1 mu = 0
class gaussian_gen(rv_continuous):
    "Gaussian distribution"
    def _pdf(self, x):
         return np.exp(-x**2 / 2.) / np.sqrt(2.0 * np.pi)

rv = gaussian_gen()
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(-5,5,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=1)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
%time

#параметры запуска: размеры выборок и количество повторов генерации
n1 = 2
n2 = 5
n3 = 7
samples_count = 1000


#массивы для сохранения средних по выборкам
a1 = np.array([])
a2 = np.array([])
a3 = np.array([])

for number in range(samples_count):
    # на каждом цикле генерируем выборки разных размеров
    sample1 = rv.rvs(size=n1)
    sample2 = rv.rvs(size=n2)
    sample3 = rv.rvs(size=n3)
    # считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
    a2 = np.append(a2, sample2.mean())
    a3 = np.append(a3, sample3.mean())
```

```python
#строим графики
#подписи для графиков
label_1 = 'sample of ' + str(n1) + ' histogram'
plt.hist(a1, bins = 50, density=True, label=label_1, alpha = 0.7 )

label_2 = 'sample of ' + str(n2) + ' histogram'
plt.hist(a2,  bins = 50, density=True, label=label_2, alpha = 0.7)

label_3 = 'sample of ' + str(n3) + ' histogram'
plt.hist(a3,  bins = 50, density=True, label=label_3, alpha = 0.7)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```
