```python
np.random.seed(33)

def normal(x, mu, sigma):
    return (1 / (sigma * math.sqrt(2 * math.pi))) * math.exp(-0.5 * ((x - mu) / sigma) ** 2 )

sigma_arr = np.random.randint(10, size = 5) 
mu_arr = np.random.randint(10, size = 5)
x_arr = np.arange(-10, 15, 0.001) 

plt.figure(figsize=(20,10))
for i,j in zip(sigma_arr, mu_arr):
    p = plt.plot(x_arr, [normal(x, i, j) for x in x_arr], 

                 label='sigma = {}, mu = {}'.format(i, j))
plt.legend()
plt.show()
```

```python
np.random.seed(33)

def exponential(x, lambda_):
    return lambda_ * math.exp(-1 * lambda_ * x)

lambda_arr = np.random.randint(50, size = 5) / 350
x_arr = np.arange(-20, 20, 0.01) 

plt.figure(figsize=(20,10))
for i in lambda_arr:
    p = plt.plot(x_arr, [exponential(x, i) for x in x_arr], 
                 label='lambda = {:.2}'.format(i))

plt.legend()
plt.show()
```
